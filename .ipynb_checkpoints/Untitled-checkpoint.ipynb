```python
def convert_string_to_number(str):
    if str.isdigit():
        return int(str)
    num1 = ''
    num2 = ''
    add = None
    for i in range(0, len(str)):
        if str[i].isdigit() and add == None:
            num1 += str[i]
        if str[i].isdigit() and add != None:
            num2 += str[i]
        if str[i] == '+':
            add = True
        if str[i] == '-':
            add = False
    try: 
        if add != None:
            if add == True:
                return int(num1) + int(num2)
            else:
                return int(num1) - int(num2)
    except: 
        return None
    return None
```

```python
def extract_data(data):
    if data.empty:
        print(f"No data found for position")
        return None
    data = data[['sofifa_id','age', 'height_cm', 'weight_kg', 'preferred_foot', 'weak_foot', 'skill_moves', 'work_rate', 'body_type', 
           'pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic', 'attacking_crossing','attacking_finishing', 
           'attacking_heading_accuracy', 'attacking_short_passing', 'attacking_volleys', 'skill_dribbling','skill_curve', 
           'skill_fk_accuracy', 'skill_long_passing', 'skill_ball_control', 'movement_acceleration', 
           'movement_sprint_speed','movement_agility', 'movement_reactions', 'movement_balance', 'power_shot_power', 
           'power_jumping','power_stamina','power_strength', 'power_long_shots', 'mentality_aggression', 
           'mentality_interceptions', 'mentality_positioning','mentality_vision', 'mentality_penalties', 
           'mentality_composure', 'defending_marking_awareness', 'defending_standing_tackle','defending_sliding_tackle']]
#     data = data.dropna()
    
#     df = data.loc[data['player_positions'].str.contains(position.upper())]
#     label_no_convert = data[position]
    df = data[['sofifa_id', 'age', 'height_cm', 'weight_kg', 'preferred_foot', 'weak_foot', 'skill_moves', 'work_rate', 'body_type', 
           'pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic', 'attacking_crossing','attacking_finishing', 
           'attacking_heading_accuracy', 'attacking_short_passing', 'attacking_volleys', 'skill_dribbling','skill_curve', 
           'skill_fk_accuracy', 'skill_long_passing', 'skill_ball_control', 'movement_acceleration', 
           'movement_sprint_speed','movement_agility', 'movement_reactions', 'movement_balance', 'power_shot_power', 
           'power_jumping','power_stamina','power_strength', 'power_long_shots', 'mentality_aggression', 
           'mentality_interceptions', 'mentality_positioning','mentality_vision', 'mentality_penalties', 
           'mentality_composure', 'defending_marking_awareness', 'defending_standing_tackle','defending_sliding_tackle']]
    
#     labels = []
    
#     for i in label_no_convert:
#         labels.append(convert_string_to_number(i))

#     labels = np.array(labels)
#     labels = pd.Series(labels, name = position)
    
#     print(labels.shape)
#     print(df.shape)

    encoder = OneHotEncoder(sparse=False) 
    one_hot_encoded = encoder.fit_transform(df[['work_rate']])

    one_hot_work_rate_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(['work_rate']))

    one_hot_encoded_body = encoder.fit_transform(df[['body_type']])

    one_hot_body_type_df = pd.DataFrame(one_hot_encoded_body, columns=encoder.get_feature_names_out(['body_type']))

    df.reset_index(drop=True, inplace=True)
    one_hot_work_rate_df.reset_index(drop=True, inplace=True)
    one_hot_body_type_df.reset_index(drop=True, inplace=True)

    # Ghép one-hot DataFrame với DataFrame gốc
    df = pd.concat([df, one_hot_work_rate_df, one_hot_body_type_df], axis=1)

    df = df.drop(['work_rate', 'body_type'], axis = 1)

    df['preferred_foot'] = df['preferred_foot'].replace({'Right': 1, 'Left': 0})
    
#     print(df.shape)

    return df
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer

# Giả sử df1 và df2 là hai DataFrame chứa chỉ số kỹ thuật của các cầu thủ
positions = ['st', 'lw', 'lf', 'cf', 'rf', 'rw', 'cam', 'lm', 
            'cm', 'rm','lwb', 'cdm', 'rwb', 'lb', 'cb', 'rb']

for i in positions:

    df1 = pd.read_csv(f'./data/player_stat_fl_pos/{i}_player.csv')
    df2 = pd.read_csv(f'./data/recommend_csv/{i}_player.csv')

    # 1. Chuẩn hóa các chỉ số kỹ thuật
    scaler = MinMaxScaler()  # Sử dụng MinMaxScaler để chuẩn hóa các chỉ số

    imputer = SimpleImputer(strategy='mean')
    # Chuẩn hóa các chỉ số kỹ thuật của df1
    df1_normalized = scaler.fit_transform(df1.iloc[:, 2:44])  # Bỏ cột ID và chuẩn hóa các chỉ số kỹ thuật

    # Chuẩn hóa các chỉ số kỹ thuật của df2
    df2_normalized = scaler.transform(df2.iloc[:, 2:44])  # Sử dụng cùng một scaler đã được fit từ df1

    # 2. Tính toán độ đo cosine giữa các cầu thủ
    cosine_sim = cosine_similarity(df1_normalized, df2_normalized)

    # Tạo DataFrame từ ma trận cosine similarity
    cosine_sim_df = pd.DataFrame(cosine_sim, index=df1['sofifa_id'], columns=df2['sofifa_id'])

    # In ra ma trận cosine similarity
    cosine_sim_df.to_csv(f'./data/cosine_player/{i}_similarity.csv')
    print(f'{i}')
```
