```python
dfs = []
    

# Lặp qua từng id và lấy dòng đầu tiên có id tương ứng
for id_ in gk:
    id_ = int(id_)  # Chuyển đổi id từ chuỗi sang số nguyên
    row = player[player['sofifa_id'] == id_].iloc[0]  # Lấy dòng đầu tiên có id trùng với id trong danh sách
    dfs.append(pd.DataFrame(row).transpose())  # Thêm dòng vào danh sách các DataFrame

# Ghép các DataFrame trong danh sách thành một DataFrame mới
result = pd.concat(dfs)
result
```
